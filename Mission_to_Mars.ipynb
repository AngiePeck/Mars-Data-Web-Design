```python
# Dependencies
from bs4 import BeautifulSoup
import requests
import pymongo
```

```python
# URL of page to be scraped
url = 'https://mars.nasa.gov/news/?page=0&per_page=40&order=publish_date+desc%2Ccreated_at+desc&search=&category=19%2C165%2C184%2C204&blank_scope=Latest'

# Retrieve page with the requests module
response = requests.get(url)
# Create BeautifulSoup object; parse with 'lxml'
soup = BeautifulSoup(response.text, 'lxml')
#print(soup.prettify())

#Retrieve title and paragraph text with variable
article_results = soup.find('div', class_='content_title')
news_title = article_results.find('a').text
news_p = soup.find('div', class_='rollover_description_inner').text

print(news_title)
print(news_p)
```

```python
#JPL Mars Space Image

from splinter import Browser

executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=True)

# URL of page to be scraped
url = 'https://www.jpl.nasa.gov/spaceimages/?search=&category=Mars'
browser.visit(url)
```

```python
from urllib.parse import urljoin
html = browser.html
soup = BeautifulSoup(html,'html.parser')
base = 'http://www.jpl.nasa.gov'

featured_image_partial_url = soup.find('a', class_='button fancybox')['data-fancybox-href']
featured_image_full_url = urljoin(base, featured_image_partial_url)
featured_image_full_url
featured_text = soup.find('a', class_="button fancybox")['data-description']
featured_text
```

```python
#Mars hemispheres
hemi_url = 'https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars'

# Retrieve page with the requests module
response = requests.get(hemi_url)
# Create BeautifulSoup object; parse with 'lxml'
soup = BeautifulSoup(response.text, 'lxml')
print(soup.prettify())
```

```python
#Retrieve the parent for the hemispheres
hemisphere_results = soup.find_all('div', class_='item')
#hemisphere_results

#Loop over hemisphere_results to find the title and image url:
for result in hemisphere_results:
    
    usgs_base = 'astropedia.astrogeology/usgs.gov'
    end = '.tif/jpg'
    title = result.find('h3').text
    img_url = result.find('img')['src']
    img_url_fullsize = result.a['href'].split("imgurl=")[0]
    
    print('*******************')
    print(title)
    print(img_url_fullsize)
    print(usgs_base + img_url_fullsize + end)
```

```python
import os.path

raw_html = ''   
if os.path.isfile('raw_html.txt'):
    with open('raw_html.txt', 'r') as file:
        raw_html = file.read()
else: 
    response = requests.get(hemi_url)
    raw_html = response.text
    with open('raw_html.txt','w') as file:
        file.write(raw_html)
```

```python
soup = BeautifulSoup(raw_html, 'lxml')

hemi_results = soup.find_all('div', class_='item')    
    
base = 'https://astrogeology.usgs.gov'

hemi_image_urls = []

for result in hemi_results:
    title = result.find('h3').text
    #print(title)
    
    image_partial_url = result.find('a', class_='itemLink')['href']
    image_full_url = urljoin(base, image_partial_url)
    #print(image_full_url)
    
    response_hemi = requests.get(image_full_url)
    soup_hemi = BeautifulSoup(response_hemi.text,'lxml')
    downloads_result = soup_hemi.find('div', class_='downloads')
    img_url = downloads_result.find('a')['href']
    #print(jpeg_url)
    
    
    hemi_image_urls.append({'title': title, 'img_url': img_url})
    
print(hemi_image_urls)
```
